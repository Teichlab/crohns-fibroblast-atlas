```python
import pandas as pd
import os
```

```python
## Set up parameters

## Genome assembly
genome='/nfs/team205/is10/resources/ucsc_genomes/hg38.fa'

## Output directory
outdir = '/lustre/scratch126/cellgen/team205/is10/fibroblasts/bulk_cytokines/cistarget/'
os.makedirs(outdir, exist_ok=True)
```

```python
## Read peaks to be used as background for analysis:
peaks = pd.read_csv('/lustre/scratch126/cellgen/team205/sk29/matthias_fb/bulk_ATACseq_pipeline/tables/peak_table_all_PCA_loadings_matched.csv')
```

```python
peaks
```

```python
## Extracting coordinates
peak_coord = peaks[['seqnames', 'start', 'end']]
```

```python
## Adding the peak's name
peak_coord['name'] = peaks['seqnames'] + ':' + peaks['start'].astype(str) + '-' + peaks['end'].astype(str)
```

```python
peak_coord
```

```python
## Export into bed format
peak_coord.to_csv((outdir + 'all_peaks.bed'), sep='\t', index=False, header=False)
```

```python
## And convert to fasta
os.system(('bedtools getfasta -fi ' + genome + ' -bed ' + outdir + 'all_peaks.bed > ' + outdir + 'all_peaks.fa'))
```
