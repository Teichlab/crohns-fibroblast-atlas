```r
#d = edges[edges$Gene %in% s5_genes]
d = edges[edges$Gene %in% cnmf_top[, 7]]

head(d)
```

```r
# Get adjacency matrix
mat = dcast(d, TF~Gene, value.var="TF2G_regulation")
mat = as.data.frame(mat)
mat[is.na(mat)] = 0
rownames(mat) = mat$TF
mat = mat[, -1]  # drop TF column
```

```r
mat
```

```r
width=2.5; height=4

options(repr.plot.width = width, repr.plot.height = height)

hc = hclust(dist(mat, method='manhattan'))
plot(as.dendrogram(hc), hang=-1, horiz=T)
tf_order = rownames(mat)[hc$order]
dev.copy(pdf, "plots/scenic_S5_regulons_row_dendrogram.pdf", width=width, height=height)  # default pdf device
dev.off()
```

```r
width=12; height=3.5
options(repr.plot.width = width, repr.plot.height = height)

hc = hclust(dist(t(mat), method='manhattan'))
plot(hc, hang=-1, cex=0.7)
gene_order = colnames(mat)[hc$order]
dev.copy(pdf, "plots/scenic_S5_regulons_column_dendrogram.pdf", width=width, height=height)  # default pdf device
dev.off()
```

```r
d$Gene = factor(d$Gene, levels=gene_order)
d$TF = factor(d$TF, levels=tf_order)
```

```r
d$TF2G_regulation = factor(d$TF2G_regulation)
```

```r
width = 12; height = 3;
options(repr.plot.width = width, repr.plot.height = height)
ggplot(d, aes(x=Gene, y=TF, fill=TF2G_importance, shape=TF2G_regulation)) +
    geom_point(size=2.5) +
    scale_fill_viridis(option="magma", direction=-1) +
    discrete_scale('shape', 'shape_d', function (n) (c(21, 23))[seq_len(n)]) +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 90, hjust=1, vjust=0.5))

ggsave("plots/scenic_S5_regulons.pdf", width=width, height=height)              
```
